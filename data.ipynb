```python
import zmail
import pandas as pd
server = zmail.server('colourgame2022@163.com','KXATNCTPLTFUFQIJ')
mails = server.get_mails(subject='colour game results',sender='colourgame2022',start_time='2022-08-03')
mails_content=[]
for mail in mails:
    mails_content.append(mail['content_text'][0].split(','))
df_data=pd.DataFrame(mails_content)
df_data.columns = ['2-2', '3-3', '1-1', '1-3', '2-1', '2-4', '4-2', '1-2','3-1','5-1','5-2','6','7']
df_data['2-3']=df_data['1-2']
df_data['3-2']=df_data['1-2']
df_data['4-1']=df_data['1-1']
df_data = df_data[['1-1', '1-2', '1-3', '2-1', '2-2', '2-3','2-4', '3-1','3-2','3-3','4-1','4-2','5-1','5-2','6','7']]
df_data
```
